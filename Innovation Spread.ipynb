## Introduction

The Lotka-Volterra equations, frequently referred to as predator-prey equations, are a set of non-lineardifferential equations constructed to describe the interaction dynamics between different species in na-ture.  Yet, since their publication many authors have proved that the applications of these equations goway beyond mathematical biology.

The Lotka-Volterra model was developed to allow ecologists to predict the potential outcome when two species are in competition for the same resources. Thus, it can also be used to analyse the outcome of competition between two competitors in a business context. Total number of potential customers is considered to be a constant limited resource. 

The Lotkaâ€“Volterra equations predict that the winner of exploitative competition for resources in stable environments should be the species with the greater K or carrying capacity, that is, the more efficient user of the resource.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
w = 4
mu = 0.5
Y, X = np.mgrid[0:w:10j, 0:w:10j]
U = X*(1-Y)
V = mu*Y*(X-1)

fig=plt.figure(figsize=(7,7))
strm=plt.streamplot(X, Y, U, V,cmap='winter', color = U,density=[3,1])
plt.title('Vector field for Lotka-Voltera model',fontsize=20)
plt.xlabel(r'$x=\frac{dR}{c}$',fontsize=18)
plt.ylabel(r'$y=\frac{bF}{a}$',fontsize=18)
plt.text(3.0,0.5,r'$\mu=0.5$',fontsize=14)
fig.savefig('lotkavoltera.png')

w = 4
mu = -0.75
Y1, X1 = np.mgrid[0:w:10j, 0:w:10j]
U1 = X1*(1-Y1)
V1 = mu*Y1*(X1-1)

fig=plt.figure(figsize=(7,7))
plt.title('Vector field for Lotka-Voltera model',fontsize=20)
plt.xlabel(r'$x=\frac{dR}{c}$',fontsize=18)
plt.ylabel(r'$y=\frac{bF}{a}$',fontsize=18)
plt.text(3.0,0.5,r'$\mu=-0.75$',fontsize=14)
strm=plt.streamplot(X1, Y1, U1, V1, cmap = 'inferno',color=U1,density=[2,1],linewidth=1)
fig.savefig('negmu.png')
```
